```python
# To load the Pandas libraries as an alias 'pds' 
import pandas as pds
# To load the numpy libraries as an alias 'npy' 
import numpy as npy 

#To create a learning model and pass it the training data
from sklearn.neighbors import KNeighborsClassifier 

from sklearn.model_selection import train_test_split 
from sklearn import metrics 

SoybeanData = pds.read_csv("soybean.csv")

df_soybean = pds.DataFrame(SoybeanData)

#Getting count of class values for stratification
Stratified_fold_1 = df_soybean[df_soybean["roots"] == 'D1']
Stratified_fold_2 = df_soybean[df_soybean["roots"] == 'D2']
Stratified_fold_3 = df_soybean[df_soybean["roots"] == 'D3']
Stratified_fold_4 = df_soybean[df_soybean["roots"] == 'D4']

#Splitting above partitions into folds for applying 5 fold cross validation
k_folds_S1 = npy.array_split(Stratified_fold_1,5)
k_folds_S2 = npy.array_split(Stratified_fold_2,5)
k_folds_S3 = npy.array_split(Stratified_fold_3,5)
k_folds_S4 = npy.array_split(Stratified_fold_4,5)

#Splitting testing and training set from obtained stratification and folds
for iterv in range(len(k_folds_S1)):
    Test_S1 = k_folds_S1[iterv]
    Test_S2 = k_folds_S2[iterv]
    Test_S3 = k_folds_S3[iterv]
    Test_S4 = k_folds_S4[iterv]
    
    frames = [Test_S1, Test_S2, Test_S3, Test_S4]
    Test_S = pds.concat(frames)
    print(Test_S)
    count=0
    Train_S1 = Train_S2 = Train_S3 = Train_S4 = [] 
    for i in range(len(k_folds_S1)):
        if iterv!=i and count!=4:
            Train_S1 += [k_folds_S1[i]]
            Train_S2 += [k_folds_S2[i]]
            Train_S3 += [k_folds_S3[i]]
            Train_S4 += [k_folds_S4[i]]
            count += 1
        else:
            continue
    Train_S = Train_S1 + Train_S2 + Train_S3 + Train_S4
    print(Train_S)

T = Train_S
x = Test_S    

T_training, T_testing, x_training, x_testing = train_test_split(T, x, test_size=0.4, random_state=1) 
        
# To train model using training set    
Model_KNN = KNeighborsClassifier(n_neighbors=3) 

# Making predictions in testing set
x_pred = Model_KNN.predict(T_testing) 

# To fit the model
Model_KNN.fit(T_training, x_training) 
#Calculating Accuracy
print("Accuracy(KNN):", metrics.accuracy_score(x_testing, x_pred))
```
